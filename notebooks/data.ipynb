```python
datasets = dict()
for dataset in ['ideal-barcode-mask-alignment', 'ru-goods-barcodes']: # 'barcode-and-qr', 'barcode-detection-annotated-dataset'
    datasets[dataset] = list()
    root, folders, files = next(os.walk(f'../data/{dataset}'))

    if dataset == 'ideal-barcode-mask-alignment':
        images_path = f'{root}/IMAGES'
        masks_path = f'{root}/MASKS'
        
        root, folders, files = next(os.walk(images_path))
        suc = 0
        fail = 0
        for f in files:
            datasets[dataset].append(
                (np.asarray(Image.open(f'{images_path}/{f}')),
                 np.asarray(Image.open(f'{masks_path}/{f}')))
                 )

    elif dataset == 'ru-goods-barcodes':
        images_path = f'{root}/images'
        datasets[dataset] = {'images': [], 'annotation': pd.read_csv(f'{root}/annotation.tsv', sep='\t')}
        root, folders, files = next(os.walk(images_path))
        for f in files:
            datasets[dataset]['images'].append(
                        plt.imread(f'{root}/{f}')
                    )
        
```
